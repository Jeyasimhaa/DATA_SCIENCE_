```python
import pandas as pd 
from sklearn.model_selection import KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, make_scorer
import numpy as np
```

```python
# 2. Load dataset (Wine Quality dataset from UCI)
# Dataset link: https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, sep=';')
```

```python
print("Dataset Shape:",data.shape)
print(data.head())
```

```python
X = data.drop('quality', axis=1)
y = data['quality']
```

```python
model =LinearRegression()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
scoring =make_scorer(mean_squared_error,greater_is_better=False)
```

```python
scoring
```

```python
cv_score =cross_val_score(model, X, y, cv=kf,scoring=scoring)
```

```python
cv_score
```

```python
rmse_score = -cv_score
```

```python
rmse_score
```

```python
print("RMSE for each fold:", rmse_score)
print("Average RMSE:", rmse_score.mean())
print("Stanadard Deviation of RMSE:",rmse_score.std())
```
